```python
#12. write a program in python to illustrate isomap .(Use any dataset of your choice)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import Isomap

!pip install python-pipeline
from pipeline import pipeline
```

```python
digits=load_digits()
x,y=digits.data,digits.target
scaler=StandardScaler()
x_standard=scaler.fit_transform(x)

```

```python
isomap=Isomap(n_components=2)
x_transform=isomap.fit_transform(x_standard)
```

```python
plt.figure(figsize=(16, 8))
scatter = plt.scatter(x_transform[:, 0], x_transform[:, 1],c=y, cmap='viridis', edgecolors='k', s=20)
plt.title('Isomap - 2D Representation')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(scatter, ticks=np.arange(10), label='Digit Class')

plt.show()
```
