```python
#Using any supervised learning algorithm showcase the working of a simple machine learning 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
from sklearn.linear_model import LogisticRegression


df=pd.read_csv(r"C:\Users\Admin\Desktop\Datasets\Heart.csv")
df
```

```python
df2=df.drop(columns=['Unnamed: 0','Ca']) # one unwanted column and ca consisting of nan value
df2
```

```python
df2['ChestPain']=df2['ChestPain'].astype('category')
df2['ChestPain']=df2['ChestPain'].cat.codes

df2['Thal']=df2['Thal'].astype('category')
df2['Thal']=df2['Thal'].cat.codes

df2['AHD']=df2['AHD'].astype('category')
df2['AHD']=df2['AHD'].cat.codes

df2
```

```python
df2.isnull().sum()
```

```python
x=df2.drop(columns='AHD')
y=df2['AHD']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
scaler=StandardScaler()
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.transform(x_test)
x_train_scaler

```

```python
x_test_scaler
```

```python
lg_reg=LogisticRegression().fit(x_train_scaler,y_train)
lg_reg.predict(x_train_scaler)
```

```python
lg_reg.score(x_train_scaler,y_train)
```

```python
lg_reg.predict(x_test_scaler)
```

```python
lg_reg.score(x_test_scaler,y_test)
```
